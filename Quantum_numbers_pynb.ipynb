```python
def quantum_numbers():
  a = int(input("put the number of n: "))
  if a<=0:
    return "the value of n shouldn't be less than 1"
  
  l_list = []
  ml_list = []
  b = a-1
  for j in range(0,a):
    l_list.append(j)

  for i in range(-b,b+1):
    ml_list.append(i)

  print("the allowed range of values l is:", l_list)
  print("the allowed range of values ml is:", ml_list)

  return 'note: the value of ms is always either 1/2 or -1/2'
```

```python
quantum_numbers()
```
